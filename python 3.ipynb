```python
class A:
    def dis(self):
        print("belongs to A")
        
class B(A):
    def dis(self):          #here we are overriding the method dis with different execution
        print("belongs to B") #method overridng is ability of a class to change the implmntation of method provided by its base class
obj = B()
obj.dis()
```
